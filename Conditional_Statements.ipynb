```python
#8 Calculate the gross salary of an employee based on basic salary
# if basic <= 10000: HRA = 20%, DA= 80%
# if basic <= 20000: HRA = 25%, DA= 90%
# otherwise: HRA = 30%, DA = 95%

salary = int(input("Enter your salary: "))
b1 = (0.2*10000) + (0.8*10000)
b2 = (0.25*20000) + (0.9*20000)

if salary <= 10000 :
  print("gross salary is: ", salary+b1)
elif salary <= 20000 :
  print("gross salary is: ", salary+b2)
else:
  print("gross salary is: ", salary+ 0.3*salary + 0.95*salary)
```

```python
#9 Calculate income tax based on annual income
# up to rps 2,50,000: no tax
# 2,50,001 - 5,00,000: 5%
# 5,00,001 - 10,00,000: 20%
# above 10,00,000: 30%

income = int(input("Enter your annual income: "))

if income <= 250000:
  tax = 0
elif income <= 500000:
  tax = (income - 250000)* 0.05
elif income <= 1000000:
  tax = (250000*0.05) + (income- 500000) * 0.20
else :
  tax = (250000*0.05) + (500000 * 0.20) + (income - 1000000) * 0.30

print("Income tax: ", tax)
```
